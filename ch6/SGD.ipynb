```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(-10,10,size=(1000,2))
def sgd_batch(batch,title):
    w=np.array([0,0])
    res=np.array([[0,0]])

    for i in range(1000):
        batchs=np.random.randint(0,1000,size=batch)
        w=w-1/(i+1)*(w-x[batchs].mean(axis=0))
        res=np.append(res,[w],axis=0)

    plt.figure(figsize=(8,5))
    plt.yticks(np.arange(-5,5,1))
    plt.title(title)
    plt.plot(res[:,0],label='x')
    plt.plot(res[:,1],label='y')
    plt.legend()
    plt.show()
```

```python
sgd_batch(1,'1-batch')
```

```python
# y=2 * X ** 2 + 3*X +5 

X=np.linspace(-10,10,1000)
Y=2 * X ** 2 + 3*X +5 
def compu(x,a,b,c):
    return a*x**2+b*x+c
def loss_fn(y,y_hat):
    return np.mean((y-y_hat)**2)
```

```python
def compute(learning_rate,batch,title):
    a=np.random.randn()
    b=np.random.randn()
    c=np.random.randn()

    loss_sum=[]
    loss=1000


    target=[]
    predict=[]
    while loss>0.01:

        batchs=np.random.randint(0,1000,size=(1,batch))
        x=X[batch]
        y=Y[batch]
        y_hat=compu(x,a,b,c)
        loss=loss_fn(y_hat,y)
      

        grad_a=(2*(y_hat-y)*x**2).mean()
        grad_b=(2*(y_hat-y)*x).mean()
        grad_c=(2*(y_hat-y)).mean()

        a-=learning_rate*grad_a
        b-=learning_rate*grad_b
        c-=learning_rate*grad_c

        
        valid_loss=loss_fn(compu(x,a,b,c),y)
        loss_sum.append(valid_loss)

        print(a,b,c)

    y_pred=compu(X,a,b,c)
    plt.figure(figsize=(8,5))
    plt.plot(X,Y,label='target',color='r')
    plt.plot(X,y_pred,label='pred',color='b')
    plt.legend()
    plt.show()


    plt.figure(figsize=(8,5))
    plt.plot(loss_sum[:150],label='loss')
    plt.show()
```
