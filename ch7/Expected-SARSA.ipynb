```python
# ä¼°è®¡action value
# Expected-SARSAç›´æ¥è®¡ç®—state value
def Expected_SARSA(gridworld,sarsaStep=5,gamma = 0.9,trajectorySteps=-1, learning_rate=0.001, final_epsilon=0.01, num_episodes=600):
    # TD error----TD target
    epsilon=0.5
    policy=np.eye(5)[np.random.randint(0,5,25)]
    q_table=np.zeros((25,5))
    state_value=np.zeros(25)

    for num in range(num_episodes):
        print("episode",f"{num}/{num_episodes}")
        # time.sleep(0.2)
        if epsilon>final_epsilon:
            epsilon-=0.001
        else:
            epsilon=final_epsilon
    
        d={1:1-epsilon*(4/5),0:epsilon/5}
        policy_epsilon=np.vectorize(d.get)(policy)

        if trajectorySteps==-1:
            stop_when_target=True
    
        Trajectory=env.getTrajectoryScore(action=np.random.randint(0,5),policy=policy_epsilon,state=10,step=500,stop_when_target=True)

        # è‡ªè½¬
        Trajectory.append((17,4,1,17,4))

        steps = len(Trajectory)
        # if steps>trajectorySteps:
        #     continue
        
        # ä¼°è®¡action value
        for step in range(steps-1,-1,-1):
            tmp_state,tmp_action,tmp_reward,next_state,next_action=Trajectory[step]


            # å¾—åˆ°expected state value
            v=tmp_reward+gamma*(q_table[next_state]*policy_epsilon[next_state]).sum()
            TD_error=q_table[tmp_state][tmp_action]-v
            q_table[tmp_state][tmp_action]-=learning_rate*TD_error
            
        policy=np.eye(5)[np.argmax(q_table,axis=1)]
        policy_epsilon=np.vectorize(d.get)(policy)
        # state value
        state_value=np.sum(policy_epsilon*q_table,axis=1)
        print('state value',state_value.reshape(5,5))
        print(env.showPolicy(policy))
```
