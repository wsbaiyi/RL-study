```python
import numpy as np
import matplotlib.pyplot as  plt
import sys
import time
from IPython.display import clear_output
sys.path.append('..')
from GridWorld_v2 import GridWorld_v2
env=GridWorld_v2(forbiddenScore=-5000, score=100,desc = [".....",".##..","..#..",".#T#.",".#..."]) 
env.show()
```

```python
# ä¼°è®¡action value
# n-stepå°±æ˜¯é‡‡æ ·stepæ­¥éª¤çš„rewardï¼Œå…¶ä½™ä¼°è®¡
def n_step_SARSA(gridworld,sarsaStep=5,gamma = 0.9,trajectorySteps=-1, learning_rate=0.001, final_epsilon=0.01, num_episodes=600):
    # TD error----TD target
    epsilon=0.5
    policy=np.eye(5)[np.random.randint(0,5,25)]
    q_table=np.zeros((25,5))
    state_value=np.zeros(25)

    for num in range(num_episodes):
        print("episode",f"{num}/{num_episodes}")
        # time.sleep(0.2)
        # clear_output(wait=True)
        if epsilon>final_epsilon:
            epsilon-=0.001
        else:
            epsilon=final_epsilon
    
        d={1:1-epsilon*(4/5),0:epsilon/5}
        policy_epsilon=np.vectorize(d.get)(policy)

        if trajectorySteps==-1:
            stop_when_target=True
    
        Trajectory=env.getTrajectoryScore(action=np.random.randint(0,5),policy=policy_epsilon,state=10,step=500,stop_when_target=True)

        # è‡ªè½¬
        for i in range(sarsaStep):
            Trajectory.append((17,4,1,17,4))

        steps = len(Trajectory)
        if steps>trajectorySteps:
            continue
        
        # ä¼°è®¡action value
        for step in range(0,len(Trajectory)-sarsaStep):
            now_state,now_action=-1,-1
            reward=0
            g=1
            for i in range(sarsaStep):
                tmp_state,tmp_action,tmp_reward,next_state,next_action=Trajectory[step+i]
                if now_state==-1:
                    now_state,now_action=tmp_state,tmp_action
                reward+=g*tmp_reward
                g*=gamma
            TD_error=q_table[now_state][now_action]-(reward+g*q_table[next_state][next_action])
            q_table[now_state][now_action]-=learning_rate*TD_error
        # å¾—åˆ°policy
        policy=np.eye(5)[np.argmax(q_table,axis=1)]
        policy_epsilon=np.vectorize(d.get)(policy)
        # state value
        state_value=np.sum(policy_epsilon*q_table,axis=1)
        print('state value',state_value.reshape(5,5))
        print(env.showPolicy(policy))
```

```python
rows = 5      #è®°å¾—è¡Œæ•°å’Œåˆ—æ•°è¿™é‡Œè¦åŒæ­¥æ”¹
columns = 5
import random
import time
def n_step_SARSA(gridworld,sarsaStep=5,epsilon=0.5,trajectorySteps=-1, learning_rate=0.001, final_epsilon=0.01, num_episodes=600, gamma = 0.9):
    """
    è¿™æ˜¯n-step-SARSAç®—æ³•

    Parameters:
    sarsaStep (int): å¯¹åº”n-step-sarsaé‚£ä¸ªnçš„å¤§å°
    trajectorySteps (int): å¯»è·¯çš„è½¨è¿¹é•¿åº¦ï¼Œå¦‚æœæ˜¯-1ï¼Œåˆ™ä¸ºå¯»åˆ°ç›®çš„åˆ™åœæ­¢ï¼Œå¦åˆ™å‚æ•°å³ä¸ºtrajectoryé•¿åº¦
    learning_rate (float): å­¦ä¹ ç‡ï¼Œç”¨äºè°ƒèŠ‚TD-target
    final_epsilon (float): epsilon-greedyçš„æ ¸å¿ƒå‚æ•°ï¼Œ0~1çš„æµ®ç‚¹æ•°ï¼Œå…¶ä¸­1åˆ™è¡¨ç¤ºå½“å‰stateæ‰€æœ‰å†³ç­–æ¦‚ç‡ä¸€æ ·ï¼Œ0åˆ™è¡¨ç¤ºå†³ç­–æ²¡æœ‰ä»»ä½•çš„éšæœºæ€§
    num_episodes (int): è¡¨ç¤ºæ¨¡å‹è¿­ä»£æ¬¡æ•°
    gamma (float): è¿‘è§†è¿œè§†ç¨‹åº¦

    Returns:
    GridWorld_v2.GridWorld_v2: æŠŠæ¨¡å‹è¿”å›å›å»
    """
    
    state_value = np.zeros((rows * columns))
    action_value = np.zeros((rows * columns, 5))
    policy = np.eye(5)[np.random.randint(0,5,size=(rows*columns))] 
    for episode in range(num_episodes):
        #æ¸…é™¤è¾“å‡ºï¼Œå¯ä»¥æ›´å¥½çš„å±•ç¤ºç­–ç•¥
        time.sleep(0.2)
        clear_output(wait=True)
        
        print("episode",f"{episode}/{num_episodes}")
        if(epsilon > final_epsilon) :
            epsilon -= 0.001
        else:
            epsilon = final_epsilon

        # p1æ˜¯ç›®æ ‡æ–¹å‘çš„æ¦‚ç‡ï¼Œp0æ˜¯å¦å¤–å››ä¸ªæ–¹å‘çš„æ¦‚ç‡
        p1 = 1-epsilon * (4/5)
        p0 = epsilon/5
        d = {1:p1, 0:p0}
        # policy_epsilonæ˜¯policyå–epsilon-greedyçš„æ¦‚ç‡å†³ç­–
        print("p1",p1,"p0",p0)
        policy_epsilon = np.vectorize(d.get)(policy)

        #cntæ•°ç»„ç”¨æ¥æ£€æŸ¥æ¯ä¸ªstateæœ‰å¤šå°‘æ¬¡è®¿é—®
        cnt = [0 for i in range(25)]
        
        initState=10
        initAction=random.randint(0,4)  

        if trajectorySteps==-1:
            stop_when_target = True
        Trajectory = env.getTrajectoryScore(action=np.random.randint(0,5),policy=policy_epsilon,state=10,step=500,stop_when_target=True)
        for i in range(sarsaStep):
            Trajectory.append((17,4,1,17,4)) #è®©ä»–æœ€åè‡ªè½¬ä¸€ä¸‹
        print("trajectorySteps",len(Trajectory))
        

        
        # æ³¨æ„è¿™é‡Œçš„è¿”å›å€¼æ˜¯å¤§å°ä¸º(trajectorySteps+1)çš„å…ƒç»„åˆ—è¡¨ï¼Œå› ä¸ºæŠŠç¬¬ä¸€ä¸ªåŠ¨ä½œä¹ŸåŠ å…¥è¿›å»äº†
        
        steps = len(Trajectory)
        if steps>trajectorySteps:
            continue
        g = 1
        reward = 0
        for k in range(0, steps, 1):
            tmpstate, tmpaction, tmpscore, nextState, nextAction  = Trajectory[k]
            reward += tmpscore * g
            g *= gamma
        print("Trajectory_score", reward)


        
        for k in range(0, steps - sarsaStep + 1, 1):
            #Stateï¼ŒActionï¼ŒRewardï¼ŒNextStateï¼ŒNextAction
            cnt[Trajectory[k][0]] += 1
            #n-step-SARSA 
            nowState,nowAction = -1,-1
            g = 1
            reward = 0
            for i in range(sarsaStep):
                #Stateï¼ŒActionï¼ŒRewardï¼ŒNextStateï¼ŒNextAction
                tmpstate, tmpaction, tmpscore, nextState, nextAction  = Trajectory[k+i]
                if nowState == -1:
                    nowState, nowAction = tmpstate, tmpaction
                reward += tmpscore * g
                g *= gamma
            reward += g*action_value[nextState][nextAction]
            
            TD_error = action_value[nowState][nowAction] - reward
            action_value[nowState][nowAction] -= learning_rate * TD_error

        # for k in range(steps - sarsaStep+1, steps, 1):
        #     # print(k)
        #     #Stateï¼ŒActionï¼ŒRewardï¼ŒNextStateï¼ŒNextAction
        #     tmpstate, tmpaction, tmpscore, nextState, nextAction  = Trajectory[k]
        #     cnt[tmpstate] += 1
        #     TD_error = action_value[tmpstate][tmpaction] - (tmpscore + gamma * action_value[nextState][nextAction])
        #     action_value[tmpstate][tmpaction] -= learning_rate * TD_error


        # policy improvement
        policy = np.eye(5)[np.argmax(action_value,axis=1)]  #qtableçš„æœ€ä¼˜å€¼ä½œä¸ºæ›´æ–°ç­–ç•¥ï¼Œå¹¶ç”¨ç‹¬çƒ­ç æ¥è¡¨ç¤º
        policy_epsilon = np.vectorize(d.get)(policy)
    
        #è¾“å‡ºæ¯ä¸ªstateçš„è®¿é—®æ¬¡æ•°
        print(np.array(cnt).reshape(5,5)) 

        state_value = np.sum(policy_epsilon * action_value,axis=1)
        mean_state_value = np.sum(policy_epsilon * action_value,axis=1).mean()
        
        gridworld.showPolicy(policy)
        print(np.round(action_value,decimals=3).reshape(-1,5))
        print(np.round(state_value,decimals=4).reshape(5,5))
        print("mean_state_value", mean_state_value)

    return gridworld
    
```
