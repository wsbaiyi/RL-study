```python
import numpy as np
import matplotlib.pyplot as  plt
import sys
sys.path.append('..')
from GridWorld_v2 import GridWorld_v2
```

```python
env=GridWorld_v2(forbiddenScore=-5000, score=100,desc = [".....",".##..","..#..",".#T#.",".#..."]) 
env.show()
```

```python
# 计算得到state value
policy=0.2*np.ones((25,5))
step=500
gamma=0.9
state_value=np.zeros(25)
learn_rate=0.01
eposides=[env.getTrajectoryScore(action=np.random.randint(0,5),policy=policy,state=np.random.randint(0,5),step=step)  for i in range(500)]
for traject in eposides:
    for step in traject:
        now_state,now_action,reward,next_state,next_action=step
        TD_error=state_value[now_state]-(reward+gamma*state_value[next_state])
        state_value[now_state]-=learn_rate*TD_error
state_value=state_value.reshape(5,5)
x = np.linspace(0,5,5)
y = np.linspace(0,5,5)

X, Y = np.meshgrid(x,y)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(Y,X,state_value, cmap='viridis')
```

```python
# 用TD 拟合state value
gamma = 0.999   #折扣因子，越接近0越近视
trajectorySteps = 20000
learning_rate = 0.01
num_episodes = 60000
epsilon=0.1
policy=np.eye(5)[np.random.randint(0,5,25)]
q_table=np.zeros((25,5))
state_value=np.zeros(25)
```

```python

# TD error----TD target
for num in range(num_episodes):
    if epsilon>0.001:
        epsilon-=0.001
    else:
        epsilon=0.001

    d={1:1-epsilon*(4/5),0:epsilon/5}
    policy_epsilon=np.vectorize(d.get)(policy)

    Trajectory=env.getTrajectoryScore(action=np.random.randint(0,5),policy=policy,state=np.random.randint(0,5),step=500)
    # 估计state value
    for step in range(len(Trajectory)-1,-1,-1):
        now_state,now_action,reward,next_state,next_action=Trajectory[step]
        TD_error=state_value[now_state]-(reward+gamma*state_value[next_state])
        state_value[now_state]-=learn_rate*TD_error
    
    # 算出action value
    for i in range(25):
        for j in range(5):
            reward,next_state=env.getNewState(action=j,state=i)
            q_table[i][j]=reward+gamma*state_value[next_state]


    # 更新policy
    policy=np.eye(5)[np.argmax(q_table,axis=1)]


    print('state value',state_value.reshape(5,5))
    print(env.showPolicy(policy))
```
