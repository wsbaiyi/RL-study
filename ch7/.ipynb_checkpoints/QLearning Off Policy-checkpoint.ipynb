```python
# ‰º∞ËÆ°optimal action value
def QLearning_Off_Policy(gridworld,sarsaStep=5,gamma = 0.9,trajectorySteps=-1, learning_rate=0.001, final_epsilon=0.01, num_episodes=600):
    # TD error----TD target
    epsilon=0.5
    policy=np.eye(5)[np.random.randint(0,5,25)]
    q_table=np.zeros((25,5))
    state_value=np.zeros(25)
    behavior_policy=0.2*np.ones((25,5))

    # off policyÔºåbehavior_policyÁîüÊàêÁ≠ñÁï•Ôºåtarget policyÁ≠ñÁï•Êõ¥Êñ∞‰∏ç‰ºöÂΩ±ÂìçÂ∑≤ÁªèÁîüÊàêÁöÑtrajectory
    if trajectorySteps==-1:
            stop_when_target=True
    poolsize=50
    TrajectPool=[env.getTrajectoryScore(action=np.random.randint(0,5),policy=behavior_policy,state=10,step=500,stop_when_target=stop_when_target) for i in range(50)]
    

    
    
    for num in range(num_episodes):
        print("episode",f"{num}/{num_episodes}")
        # time.sleep(0.2)
        if epsilon>final_epsilon:
            epsilon-=0.001
        else:
            epsilon=final_epsilon
    
        d={1:1-epsilon*(4/5),0:epsilon/5}
        policy_epsilon=np.vectorize(d.get)(policy)
        for Trajectory in TrajectPool:
        # # Ëá™ËΩ¨
        # Trajectory.append((17,4,1,17,4))

            steps = len(Trajectory)
            for step in range(steps-1,-1,-1):
                tmp_state,tmp_action,tmp_reward,next_state,next_action=Trajectory[step]
    
                # q-learning ÂæóÂà∞
                TD_error=q_table[tmp_state][tmp_action]-(tmp_reward+gamma*q_table[next_state].max())
                q_table[tmp_state][tmp_action]-=learning_rate*TD_error
            
        policy=np.eye(5)[np.argmax(q_table,axis=1)]
        policy_epsilon=np.vectorize(d.get)(policy)
        # state value
        state_value=np.sum(policy_epsilon*q_table,axis=1)
        print('state value',state_value.reshape(5,5))
        print(env.showPolicy(policy))
```
