```python
import numpy as np     #蜿ｪ髴隕∽ｸ玖ｽｽnumpy蠎灘叉蜿ｯ
import random
import sys
sys.path.append('..')
import GridWorld_v3
from draw import draw  #扈伜崟蜃ｽ謨ｰ
from ExperienceReplayBuffer import ExperienceReplayBuffer
import tqdm
```

```python
rows = 5      #隶ｰ蠕苓｡梧焚蜥悟玲焚霑咎㈹隕∝酔豁･謾ｹ
columns = 5
gridworld = GridWorld_v3.GridWorld_v3(forbiddenAreaScore=-10, score=1,desc = [".....",".##..","..#..",".#T#.",".#..."]) 
gridworld.show()
trajectorySteps = 1000
```

```python
state_values = np.array([random.random() for i in range(25)])
learning_rate = 0.01
gamma = 0.99
```

```python
import torch.nn as nn
import torch
import torch.nn.functional as F
import torch.optim as optim
class model(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.f1=nn.Linear(2,64)
        self.f2=nn.Linear(64,128)
        self.f3=nn.Linear(128,1)
    def forward(self,data):
        data=F.relu(self.f1(data))
        data=F.relu(self.f2(data))
        return self.f3(data)
```

```python
experiencePool = ExperienceReplayBuffer(max_size=10000)
policy = 0.2 * np.ones((rows*columns,5))
episodes = [gridworld.getTrajectoryScore(nowState = 10,action = random.randint(0,4),policy = policy) 
            for i in range(500)]
for trajectory in episodes:
    for exp in trajectory:
        experiencePool.add_expericence(exp)
```

```python
lr = 0.001
network = model()
optimizer = optim.Adam(network.parameters(), lr = lr)
batch_size = 100
```

```python
batch = experiencePool.sample_batch(batch_size=batch_size)
tmpstate, tmpaction, tmpscore, nextState, nextAction, terminal = batch
```

```python
tmpstate=tmpstate.reshape(batch_size,1)
# 蠕怜芦陦悟礼ｼ門捷
tmpstate=np.hstack((tmpstate//5,tmpstate%5))
tmpstate[:5]
```

```python
for _ in tqdm.tqdm(range(100000)):
    
    optimizer.zero_grad()
    
    batch = experiencePool.sample_batch(batch_size=batch_size)
    tmpstate, tmpaction, tmpscore, nextState, nextAction, terminal = batch

    tmpstate = tmpstate.reshape(batch_size,1)
    tmpstate = np.hstack((tmpstate // 5, tmpstate % 5))  #蟆state蜿俶蝉ｺ陦悟予n    nextState = nextState.reshape(batch_size,1)
    nextState = np.hstack((nextState // 5, nextState % 5))  #蟆state蜿俶蝉ｺ陦悟予n
    tmpstate = torch.tensor(tmpstate,dtype = torch.float32).view(batch_size,2)
    nextState = torch.tensor(nextState,dtype = torch.float32).view(batch_size,2)

    tmpaction = torch.tensor(tmpaction,dtype = torch.int).view(batch_size,1)
    nextAction = torch.tensor(nextAction,dtype = torch.int).view(batch_size,1)

    tmpscore = torch.tensor(tmpscore,dtype = torch.float32).view(batch_size,1)
    terminal = torch.tensor(terminal,dtype = torch.float32).view(batch_size,1)

    # 蠕怜芦target 
    with torch.no_grad():
        next_state_value=network(nextState)
        target=tmpscore+(1-terminal)*gamma*next_state_value
    pred_value=network(tmpstate)
    loss=F.mse_loss(target,pred_value)
    loss.backward()
    optimizer.step()


    if _%2000==0:
        state=np.array([i for i in range(25)])
        state=np.hstack((state//5,state%5))
        state=torch.tensor(state,dtype=torch.float32).reshape(25,2)
        output=network(state).view(-1).detach().numpy()
        action_value=np.zeros((25,5))
        for i in range(25):
            for j in range(5):
                score,next_state=gridworld.getScore(i,j)
                action_value[i][j]=score+gamma*output[next_state]

        p=np.argmax(action_value,axis=1)
        value=np.max(action_value,axis=1)

        draw(value.reshape(5,5),p)
    
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

class model_2(nn.Module):
    def __init__(self):
        super(model_2, self).__init__()
        self.fc1 = nn.Linear(2,128)
        self.fc2 = nn.Linear(128,64)
        self.fc3 = nn.Linear(64,5)  #諡溷粋逧Бction valueｼ5荳ｪ霎灘ｺ
    def forward(self, state):
        h1 = self.fc1(state)
        h1 = F.relu(h1)
        h2 = self.fc2(h1)
        h2 = F.relu(h2)
        Q = self.fc3(h2)
        return Q
        
```

```python
lr = 0.0015
network = model_2()
optimizer = optim.Adam(network.parameters(), lr = lr)
batch_size = 100
```

```python
for _ in tqdm.tqdm(range(100000)):
    
    optimizer.zero_grad()
    
    batch = experiencePool.sample_batch(batch_size=batch_size)
    tmpstate, tmpaction, tmpscore, nextState, nextAction, terminal = batch

    tmpstate = tmpstate.reshape(batch_size,1)
    tmpstate = np.hstack((tmpstate // 5, tmpstate % 5))  #蟆state蜿俶蝉ｺ陦悟予n    nextState = nextState.reshape(batch_size,1)
    nextState = np.hstack((nextState // 5, nextState % 5))  #蟆state蜿俶蝉ｺ陦悟予n
    tmpstate = torch.tensor(tmpstate,dtype = torch.float32).view(batch_size,2)
    nextState = torch.tensor(nextState,dtype = torch.float32).view(batch_size,2)

    tmpaction = torch.tensor(tmpaction,dtype = torch.int).view(batch_size,1)
    nextAction = torch.tensor(nextAction,dtype = torch.int).view(batch_size,1)

    tmpscore = torch.tensor(tmpscore,dtype = torch.float32).view(batch_size,1)
    terminal = torch.tensor(terminal,dtype = torch.float32).view(batch_size,1)

    # 蠕怜芦target 
    with torch.no_grad():
        next_state_value=network(nextState)
        b=[[next_state_value[i,nextAction[i]]] for i in range(batch_size)]
        next_state_value=torch.tensor(b,dtype=torch.float32)
        target=tmpscore+(1-terminal)*gamma*next_state_value

    
    pred_value=network(tmpstate)
    target=[[pred_value[i,tmpaction[i]]] for i in range(batch_size)]
    target=torch.tensor(target,dtype=torch.float32).reshape(-1,1)

    loss=F.mse_loss(target,pred_value)
    loss.backward()
    optimizer.step()


    if _%2000==0:
        state=np.array([i for i in range(25)])
        state=np.hstack((state//5,state%5))
        state=torch.tensor(state,dtype=torch.float32).reshape(25,2)

        # 逾樒ｻ冗ｽ醍ｻ懃峩謗･鬚豬蟻ction value
        action_value=network(state).view(-1,5).detach().numpy()

        value=np.sum(action_value*policy,axis=1)
        p=np.argmax(action_value,axis=1)
        value=np.max(action_value,axis=1)

        draw(value.reshape(5,5),p)
    
```

```python
# DQN 逕ｨ逾樒ｻ冗ｽ醍ｻ懷ｾ怜芦optimal action value
for _ in tqdm.tqdm(range(100000)):
    
    optimizer.zero_grad()
    
    batch = experiencePool.sample_batch(batch_size=batch_size)
    tmpstate, tmpaction, tmpscore, nextState, nextAction, terminal = batch

    tmpstate = tmpstate.reshape(batch_size,1)
    tmpstate = np.hstack((tmpstate // 5, tmpstate % 5))  #蟆state蜿俶蝉ｺ陦悟予n    nextState = nextState.reshape(batch_size,1)
    nextState = np.hstack((nextState // 5, nextState % 5))  #蟆state蜿俶蝉ｺ陦悟予n
    tmpstate = torch.tensor(tmpstate,dtype = torch.float32).view(batch_size,2)
    nextState = torch.tensor(nextState,dtype = torch.float32).view(batch_size,2)

    tmpaction = torch.tensor(tmpaction,dtype = torch.int).view(batch_size,1)
    nextAction = torch.tensor(nextAction,dtype = torch.int).view(batch_size,1)

    tmpscore = torch.tensor(tmpscore,dtype = torch.float32).view(batch_size,1)
    terminal = torch.tensor(terminal,dtype = torch.float32).view(batch_size,1)

    # 蠕怜芦target 
    with torch.no_grad():
        next_state_value=network(nextState)
        # max蠕怜芦逧譏ｯ荳荳ｪ蜈扈Ыn        c=torch.max(next_state_value,dim=1)[0]
        # break
        next_state_value=torch.tensor(c,dtype=torch.float32).reshape(batch_size,1)
        target=tmpscore+(1-terminal)*gamma*next_state_value

    
    pred_value=network(tmpstate)
    target=[[pred_value[i,tmpaction[i]]] for i in range(batch_size)]
    target=torch.tensor(target,dtype=torch.float32).reshape(-1,1)

    loss=F.mse_loss(target,pred_value)
    loss.backward()
    optimizer.step()


    if _%2000==0:
        state=np.array([i for i in range(25)])
        state=np.hstack((state//5,state%5))
        state=torch.tensor(state,dtype=torch.float32).reshape(25,2)

        # 逾樒ｻ冗ｽ醍ｻ懃峩謗･鬚豬蟻ction value
        action_value=network(state).view(-1,5).detach().numpy()

        value=np.sum(action_value*policy,axis=1)
        p=np.argmax(action_value,axis=1)
        value=np.max(action_value,axis=1)

        draw(value.reshape(5,5),p)
    
```
